```python
import random

# Function to generate a random number and populate the box
def initialize_box():
    size = random.randint(1, 20)  # Random size of the box
    box = []
    type_1 = random.randint(0, size)
    for _ in range(type_1):
        box.append('A')
    type_2 = random.randint(0, size - type_1)
    for _ in range(type_2):
        box.append('B')
    type_3 = size - type_1 - type_2
    for _ in range(type_3):
        box.append('C')
    return box

# Function to randomly choose pairs and update the box until equilibrium
def update_box(box):
    boxes = []  # List to store boxes from each iteration
    while True:
        # Append the current box to the list
        boxes.append(box.copy())
        
        # Check if all elements in the box are the same
        if len(set(box)) == 1:
            #print('You have reached the redundant equilibrium stage.')
            break
        
        # Randomly choose two distinct indices
        index1, index2 = random.sample(range(len(box)), 2)
        
        # Replace the pair based on the rules
        if (box[index1], box[index2]) in [('A', 'B'), ('B', 'A')]:
            # Replace the pair with 'C'
            #print('If the particles',box[index1], 'and', box[index2], 'were to collide, we are left with the box:' )
            box[index1] = 'C'
            del box[index2]
            #print(box)
        elif (box[index1], box[index2]) in [('A', 'C'), ('C', 'A')]:
            # Replace the pair with 'B'
            #print('If the particles',box[index1], 'and', box[index2], 'were to collide, we are left with the box:' )
            box[index1] = 'B'
            del box[index2]
            #print(box)
        elif (box[index1], box[index2]) in [('B', 'C'), ('C', 'B')]:
            # Replace the pair with 'A'
            #print('If the particles',box[index1], 'and', box[index2], 'were to collide, we are left with the box:' )
            box[index1] = 'A'
            del box[index2]
            #print(box)

    return boxes

length = {}
for i in range(1, 150):
    # Initialize the box
    box = initialize_box()
    #print("Initial Box:", box)

    # Update the box until equilibrium
    updated_boxes = update_box(box.copy())
    #print('The final box is:', updated_boxes[-1])
    final = len(updated_boxes[-1])
    if final in length:
        length[final] += 1
    else:
        length[final] = 1

print("Length Dictionary:", length)
    
# Convert the dictionary into lists for plotting
size_final_box = list(length.keys())
frequency = list(length.values())


# Plotting
plt.bar(size_final_box, frequency)
plt.xlabel('Size of redundant box')
plt.xticks(range(min(size_final_box), max(size_final_box)+1, 1))
plt.ylabel('Frequency')
plt.title('Frequency of different sized final boxes')
plt.show()
```
